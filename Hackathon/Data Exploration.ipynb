```python
order_sg = order[order["Billing Country"] == "SG"]
order_sg = order_sg.reset_index(drop=True)
previous_status = None
for row in range(order_sg.shape[0]):
    if pd.isnull(order_sg.loc[row, 'Fulfillment Status']):
        order_sg.loc[row, "Fulfillment Status"] = previous_status
    else:
        previous_status = order_sg.loc[row, "Fulfillment Status"]
order_sg
```

## Identify Same Order

```python
order["ID"] = 0
```

```python
previous_email = None
order_no = 0
for row in range(order.shape[0]):
    if order.loc[row,'Email'] != previous_email:
        previous_email = order.loc[row,'Email']
        order_no += 1
    order.loc[row, 'ID'] = order_no
```

```python
order
```

## Create Basket

```python
basket = (order
          .groupby(['ID', 'Lineitem name'])['Lineitem quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('ID'))
```

```python
basket.shape
```

## One-hot Encoding

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

## Association

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules.sort_values(['lift'], ascending=False)
```

```python
rules.to_csv('Association_result.csv')
```
