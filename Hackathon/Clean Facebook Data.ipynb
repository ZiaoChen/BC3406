# Clean Facebook Data

## Import Packages

```python
import pandas as pd
import numpy as np
```

## Read Data

```python
facebook = pd.read_csv('facbook.csv')
facebook
```

## Get Time

```python
facebook['Year'] = '2018'
facebook['Month'] = '01'
facebook["Date"] = '01'
from datetime import datetime
def get_time(row):
    row["Year"] = int(row["created_time"][:4])
    row["Month"] = int(row["created_time"][5:7])
    row["Date"] = int(row["created_time"][8:10])
    return row

facebook = facebook.apply(get_time, axis=1)
facebook
```

## Get Link Content

```python
facebook['no_link'] = 0
facebook['youtube_link'] = 0
facebook['facebook_link'] = 0
facebook['paula_link'] = 0
facebook['twitter_link'] = 0
facebook['other_link'] = 0
def get_link_content(row):
    if pd.isnull(row['link']):
        row['no_link'] = 1
    elif 'youtube' in row['link']:
        row['youtube_link'] = 1
    elif 'facebook' in row['link']:
        row['facebook_link'] = 1
    elif 'paulaschoice' in row['link']:
        row['paula_link'] = 1
    elif 'twitter' in row['link']:
        row['twitter_link'] = 1
    else:
        row['other_link'] = 1
    return row
facebook = facebook.apply(get_link_content, axis=1)
facebook
```

## Message Processing

```python
facebook = facebook[pd.notnull(facebook['message'])]
```

```python
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

remove_list = ['day', 'oil', 'paulaschoice', 'sg','u', 'http', 'know', 'using','nan']
def tokenize(text):

    #Create Stemmer
    #stemmer = PorterStemmer()
    stemmer = WordNetLemmatizer()

    #Remove irrelevant character
    text = re.sub(r"[^a-zA-Z]", ' ', text)

    #Tokenization
    tokens = word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation and len(i)>1]

    #Stemming
    stems = stem_tokens(tokens, stemmer)
    return stems

#Stemming Function`
def stem_tokens(t,s):
    stemmed=[]
    for item in t:
        # stemmed.append(s.stem(item))
        if s.lemmatize(item) not in remove_list:
            stemmed.append(s.lemmatize(item))
    return stemmed

max_feature = 20
def combine_title_message(row):
    return str(row['message']) + str(row['name'])

tfidf_vectorizer = TfidfVectorizer(stop_words='english', tokenizer=tokenize, max_features = max_feature)
tfidf_data = tfidf_vectorizer.fit_transform(facebook.apply(combine_title_message, axis=1))
```

## Get Length of Message

```python
def get_length(msg):
    return len(msg.split(" "))

facebook["number_of_words"] = facebook['message'].apply(get_length)
```

## Drop Unused Columns

```python
unused_column = ['id', 'name', 'created_time', 'link', 'message', 'day', 'Date']
facebook = facebook[[i for i in facebook.columns if i not in unused_column]]
facebook
```
